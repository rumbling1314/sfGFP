```python
import csv
import torch
from esm.models.esm3 import ESM3
from esm.sdk.api import ESMProtein, GenerationConfig

# ===================== 参数设置（按需修改） =====================
# sfGFP 野生型序列（238aa）
WT_SEQUENCE = "MSKGEELFTGVVPILVELDGDVNGHKFSVSGEGEGDATYGKLTLKFICTTGKLPVPWPTLVTTFSYGVQCFSRYPDHMKQHDFFKSAMPEGYVQERTIFFKDDGNYKTRAEVKFEGDTLVNRIELKGIDFKEDGNILGHKLEYNYNSHNVYIMADKQKNGIKVNFKIRHNIEDGSVQLADHYQQNTPIGDGPVLLPDNHYLSTQSALSKDPNEKRDHMVLLEFVTAAGITLGMDELYK"

# 突变区间（生物实验 1 起始编号，包含首尾）
MUT_START_1 = 196
MUT_END_1   = 203

NUM_GENERATE  = 500    # 生成总条数
TEMPERATURE   = 0.8    # 多样性：0.6保守 / 0.8平衡 / 1.1高多样性
NUM_STEPS     = 8      # 解码步数，8步兼顾速度与质量

OUTPUT_FASTA = "sfgfp_196-203_mutants.fasta"
OUTPUT_CSV   = "sfgfp_196-203_mutants.csv"
# ==============================================================

# 坐标转换 + 构建掩码模板
mut_start = MUT_START_1 - 1
mut_end   = MUT_END_1 - 1
seq_len   = len(WT_SEQUENCE)
region_len = mut_end - mut_start + 1

wt_list = list(WT_SEQUENCE)
prompt_template = wt_list.copy()
for i in range(mut_start, mut_end + 1):
    prompt_template[i] = "_"
prompt_seq = "".join(prompt_template)

# 加载模型（首次运行自动下载权重，约 2.8GB）
print("▶ 正在加载 ESM3 1.4B 模型...")
model = ESM3.from_pretrained("esm3-sm-open-v1").to("cuda")
model = model.half()  # 半精度推理，显存占用减半
print(f"✅ 模型加载完成")
print(f"🧬 突变区间：第 {MUT_START_1}-{MUT_END_1} 号氨基酸，共 {region_len} 个位点")
print(f"🚀 开始生成 {NUM_GENERATE} 条变体...\n")

all_mutants = []

for idx in range(NUM_GENERATE):
    # 调用模型补全掩码
    protein = ESMProtein(sequence=prompt_seq)
    protein = model.generate(
        protein,
        GenerationConfig(
            track="sequence",
            num_steps=NUM_STEPS,
            temperature=TEMPERATURE
        )
    )
    mutant_seq = protein.sequence

    # 严格校验：无终止密码子 + 长度正确 + 非突变区完全锁死
    valid = True
    if "*" in mutant_seq or len(mutant_seq) != seq_len:
        valid = False
    else:
        for i in range(seq_len):
            if mut_start <= i <= mut_end:
                continue
            if mutant_seq[i] != WT_SEQUENCE[i]:
                valid = False
                break

    if valid:
        mut_count = sum(
            1 for i in range(mut_start, mut_end + 1)
            if mutant_seq[i] != WT_SEQUENCE[i]
        )
        all_mutants.append({
            "id": f"esm3_mut_{idx+1:04d}",
            "num_mutations": mut_count,
            "sequence": mutant_seq
        })

    # 进度打印
    if (idx + 1) % 100 == 0:
        print(f"进度：{idx+1}/{NUM_GENERATE} | 有效序列：{len(all_mutants)} 条")

# 序列去重
unique_mutants = []
seen = set()
for mut in all_mutants:
    if mut["sequence"] not in seen:
        seen.add(mut["sequence"])
        unique_mutants.append(mut)

print(f"\n🎉 生成完成！")
print(f"有效序列：{len(all_mutants)} 条 | 去重后：{len(unique_mutants)} 条")

# 保存结果
with open(OUTPUT_FASTA, "w") as f:
    for mut in unique_mutants:
        f.write(f">{mut['id']}_mut{mut['num_mutations']}\n{mut['sequence']}\n")

with open(OUTPUT_CSV, "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=["id", "num_mutations", "sequence"])
    writer.writeheader()
    writer.writerows(unique_mutants)

print(f"\n📁 结果文件已保存：")
print(f"  - {OUTPUT_FASTA}")
print(f"  - {OUTPUT_CSV}")
```
